On vient d'obtenir le décompte des valeurs uniques dans la colonne 'Salary Estimate' du DataFrame.

Création de nouvelles colonnes (salaire minimum, salaire maximum, salaire moyen)

```python
df['min_salary'] = 0
df['max_salary'] = 0
df['avg_salary'] = 0

for i in range(len(df)):
    try:
        df.loc[i, "min_salary"] = int(df['Salary Estimate'][i].split(" ")[0].split("-")[0].replace("$", "").replace("K", ""))
        df.loc[i, "max_salary"] = int(df['Salary Estimate'][i].split(" ")[0].split("-")[1].replace("$", "").replace("K", ""))
    except:
        df.loc[i, "min_salary"] = int(df['Salary Estimate'][i].split("(E")[0].split("-")[0].replace("$", "").replace("K", ""))
        df.loc[i, "max_salary"] = int(df['Salary Estimate'][i].split("(E")[0].split("-")[1].replace("$", "").replace("K", ""))
    finally:
        df.loc[i, "Salary Estimate"] = str(df.loc[i, "min_salary"]) + "-" + str(df.loc[i, "max_salary"])
        df.loc[i, "avg_salary"] = np.mean([df.loc[i, "min_salary"], df.loc[i, "max_salary"]])

df.head()
```

Interprétation du résultat:

1 df['min_salary']=0, df['max_salary']=0, df['avg_salary']=0: Ces instructions créent trois nouvelles colonnes 'min_salary', 'max_salary' et 'avg_salary' dans le DataFrame df et les initialisent toutes à 0.

2 La boucle for i in range(len(df)): parcourt chaque ligne du DataFrame pour effectuer les calculs.

3 Le bloc try-except permet de gérer les deux formats possibles de la colonne 'Salary Estimate'.

a. Dans le premier cas (dans le bloc try), la valeur de la colonne 'Salary Estimate' est supposée être de la forme 'XXK-YYK' (par exemple '50K-80K'). Le code extrait les valeurs 'XX' et 'YY', enlève les symboles '$' et 'K', puis convertit les chaînes en entiers pour obtenir les salaires minimum et maximum.

b. Dans le deuxième cas (dans le bloc except), la valeur de la colonne 'Salary Estimate' est supposée être de la forme 'XXK (E' (par exemple '80K (Est)'). Le code extrait la valeur 'XX', enlève les symboles '$' et 'K', puis convertit la chaîne en entier pour obtenir le salaire minimum. La valeur maximale reste 0 car il n'y a pas d'information disponible.

4 Enfin, dans le bloc finally, les nouvelles valeurs de salaire minimum et maximum sont utilisées pour mettre à jour la colonne 'Salary Estimate'. La colonne 'avg_salary' est remplie avec la moyenne des valeurs de salaire minimum et maximum.

Extraction des Jobs Description:

```python
def extract_keywords_fr(description):
    keywords = []
    if '#python' in description.lower():
        keywords.append('Python')
    if '#r studio' in description.lower():
        keywords.append('R Studio')
    if '#spark' in description.lower():
        keywords.append('Spark')
    if '#aws' in description.lower():
        keywords.append('AWS')
    if '#excel' in description.lower():
        keywords.append('Excel')
    return keywords
df['Mots Clés'] = df['Job Description'].apply(extract_keywords_fr)
print(df[['Job Description', 'Mots Clés']])
```

```python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['spark_yn'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```
